```python
from ultralytics import YOLO
import os
```

```python
from google.colab import drive

drive.mount("/content/gdrive")
```

```python
# создаем необходимые папки
RootDir="/content/gdrive/My Drive/MenOnCamera/"

files=os.listdir(RootDir)
if "DataSet" not in files:
  os.mkdir(os.path.join(RootDir,"DataSet"))
  os.mkdir(os.path.join(RootDir,"DataSet","images"))
  os.mkdir(os.path.join(RootDir,"DataSet","images","train"))
  os.mkdir(os.path.join(RootDir,"DataSet","images","val"))

  os.mkdir(os.path.join(RootDir,"DataSet","labels"))
  os.mkdir(os.path.join(RootDir,"DataSet","labels","train"))
  os.mkdir(os.path.join(RootDir,"DataSet","labels","val"))
```

```python
# разделение файлов на тренироваочные и валидоционные
def AssembleDataset(k=0.2):
  imagParh=os.path.join(RootDir,"DataSet","images","train")
  labelsPath=os.path.join(RootDir,"DataSet","labels","train")
  images=os.listdir(imagParh)
  labels=os.listdir(labelsPath)

  total=len(labels)
  cutnum=1//k #число показывает в каком соотношении делить датасет

  num=0
  for label in labels:
    name=label[:-4]+".jpg"
    print(name)
    if name in images:
      if num%cutnum==0:
        print(name)
        os.replace(os.path.join(labelsPath,label),os.path.join(RootDir,"DataSet","labels","val",label))
        os.replace(os.path.join(imagParh,name),os.path.join(RootDir,"DataSet","images","val",name))

    num+=1
```

```python
AssembleDataset()
```

```python
# удаление мусора
trashpath="/content/gdrive/My Drive/"
files=os.listdir(trashpath)
for i in files:
  if i[-4:]==".txt":
    print(i)
    os.remove(os.path.join(trashpath,i))
```

```python
# теринровка модели
model = YOLO("yolov8x.yaml")  
print(model.info())
results = model.train(data=os.path.join(RootDir, "config.yaml"), epochs=1)
```

```python
# проверка модели на валедоционных данных
metrics = model.val()  
metrics.box.map    
metrics.box.map50  
metrics.box.map75  
metrics.box.maps   
```
